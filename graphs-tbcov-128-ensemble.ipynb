```python
import pickle as pkl

with open("/kaggle/input/tbcov-ae-128/dicionario",'rb') as file:
    dicionario = pkl.load(file)
    file.close()
    
with open("/kaggle/input/tbcov-ae-128/dicionario_encoded",'rb') as file:
    dicionario_encoded = pkl.load(file)
    file.close()
```

```python
list_pd = []
for key in dicionario_encoded.keys():
    if(key == "LinearDiscr"): continue
    dic_pd = {}
    dic_pd["encoded"] = "sim"
    dic_pd["model"] = key
    dic_pd["fit_time"] = float("%.3f"%dicionario_encoded[key]['fit_time'])
    dic_pd["score_time"] = float("%.3f"%dicionario_encoded[key]['score_time'])
    dic_pd["train_acc"] = float("%.3f"%dicionario_encoded[key]['train_acc'])
    dic_pd["test_acc"] = float("%.3f"%dicionario_encoded[key]['test_acc'])
    dic_pd["train_prec_weighted"] = float("%.3f"%dicionario_encoded[key]['train_prec_weighted'])
    dic_pd["test_prec_weighted"] = float("%.3f"%dicionario_encoded[key]['test_prec_weighted'])
    dic_pd["train_rec_weighted"] = float("%.3f"%dicionario_encoded[key]['train_rec_micro'])
    dic_pd["test_rec_weighted"] = float("%.3f"%dicionario_encoded[key]['test_rec_micro'])
    dic_pd["train_AP"] = float("%.3f"%dicionario_encoded[key]['train_AP'])
    dic_pd["test_AP"] = float("%.3f"%dicionario_encoded[key]['test_AP'])
    list_pd.append(dic_pd)
```

```python
for key in dicionario.keys():
    if(key == "LinearDiscr"): continue
    dic_pd = {}
    dic_pd["encoded"] = "nao"
    dic_pd["model"] = key
    dic_pd["fit_time"] = float("%.3f"%dicionario[key]['fit_time'])
    dic_pd["score_time"] = float("%.3f"%dicionario[key]['score_time'])
    dic_pd["train_acc"] = float("%.3f"%dicionario[key]['train_acc'])
    dic_pd["test_acc"] = float("%.3f"%dicionario[key]['test_acc'])
    dic_pd["train_prec_weighted"] = float("%.3f"%dicionario[key]['train_prec_weighted'])
    dic_pd["test_prec_weighted"] = float("%.3f"%dicionario[key]['test_prec_weighted'])
    dic_pd["train_rec_weighted"] = float("%.3f"%dicionario[key]['train_rec_micro'])
    dic_pd["test_rec_weighted"] = float("%.3f"%dicionario[key]['test_rec_micro'])
    dic_pd["train_AP"] = float("%.3f"%dicionario[key]['train_AP'])
    dic_pd["test_AP"] = float("%.3f"%dicionario[key]['test_AP'])
    list_pd.append(dic_pd)
```

```python
import pandas as pd
df_c = pd.DataFrame(list_pd, columns=["encoded","model","fit_time","score_time","train_acc","test_acc","train_prec_weighted","test_prec_weighted","train_prec_weighted","test_rec_weighted","train_AP","test_AP"])
df_c.sort_values("model", inplace=True)
df_c
```

```python
baggingHistory = pd.read_csv("/kaggle/input/ensenble-tbcov-128/bagging_svc_n_votting_treino_costs.csv")
baggingHistoryAE = pd.read_csv("/kaggle/input/ensenble-tbcov-128/bagging_s_treino_costs.csv")
```

```python
import matplotlib.pyplot as plt
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_figheight(5)
fig.set_figwidth(15)

ax1.plot(baggingHistoryAE["time"],baggingHistoryAE["memory_percent"],label="AE_Bagging")
ax1.plot(baggingHistory["time"],baggingHistory["memory_percent"],label="Bagging")

ax1.set_xlabel("time")
ax1.set_ylabel("percent memory")
ax1.legend(loc='best')

ax3.plot(baggingHistoryAE["time"],baggingHistoryAE["cpu_percent"],label="AE_Stacked")
ax3.plot(baggingHistory["time"],baggingHistory["cpu_percent"],label="Stacked")

ax3.set_xlabel("time")
ax3.set_ylabel("percent CPU")
ax3.legend(loc='best');

ax2.plot(adaHistoryAE["time"],adaHistoryAE["memory_percent"],label="AE_AdaBoost")
ax2.plot(adaHistory["time"],adaHistory["memory_percent"],label="AdaBoost")

ax2.set_xlabel("time")
ax2.set_ylabel("percent memory")
ax2.legend(loc='best')

ax4.plot(adaHistoryAE["time"],adaHistoryAE["cpu_percent"],label="AE_AdaBoost")
ax4.plot(adaHistory["time"],adaHistory["cpu_percent"],label="AdaBoost")

ax4.set_xlabel("time")
ax4.set_ylabel("percent CPU")
ax4.legend(loc='best');
```
